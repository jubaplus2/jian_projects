```python
import pandas as pd
import json
import numpy as np
```

```python
actual_20181010_11=json.load(open("/home/jian/Projects/Big_Lots/Weather/Hurricane/json_data/daily/2018-10-10: 11.json","r"))
actual_20181010_14=json.load(open("/home/jian/Projects/Big_Lots/Weather/Hurricane/json_data/daily/2018-10-10: 14.json","r"))
actual_20181010_17=json.load(open("/home/jian/Projects/Big_Lots/Weather/Hurricane/json_data/daily/2018-10-10: 17.json","r"))
```

```python
pd_1010_11=pd.DataFrame()
count_k=0
for zip_cd in actual_20181010_11.keys():
    count_k+=1
    reponse_of_zip=actual_20181010_11[zip_cd]
    weather_of_zip_list=reponse_of_zip['weather']
    
    if 'deg' in reponse_of_zip['wind']:
        wind_deg=reponse_of_zip['wind']['deg']
    else:
        wind_deg=np.nan
    
    if 'gust' in reponse_of_zip['wind']:
        wind_gust=reponse_of_zip['wind']['gust']
    else:
        wind_gust=np.nan

    wind_speed=reponse_of_zip['wind']['speed']        
        
    for j in range(len(weather_of_zip_list)):
        if weather_of_zip_list[j]['main']=="Rain":
            main="Rain"
            discription=weather_of_zip_list[j]['description']
            if 'rain' in reponse_of_zip.keys():
                rain_volume=reponse_of_zip['rain']
        
            df=pd.DataFrame({"Zip":zip_cd,"Main":main,"Discription":discription,"Rain_Volume":rain_volume,
                            "wind_deg":wind_deg,"wind_speed":wind_speed},index=[count_k])
            pd_1010_11=pd_1010_11.append(df)
```
