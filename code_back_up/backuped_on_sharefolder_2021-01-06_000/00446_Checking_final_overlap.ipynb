```python
df_TA_view=file_excel.parse("view_by_TA",dtype=str)
df_TA_view.head(2)
```

```python
df_TA_view['trans_P_zips']=df_TA_view['trans_P_zips'].apply(eval)
df_TA_view['trans_S_zips']=df_TA_view['trans_S_zips'].apply(eval)

df_TA_view['TA_zip_list']=df_TA_view['trans_P_zips']+df_TA_view['trans_S_zips']
df_TA_view['TA_zip_list']=df_TA_view['TA_zip_list'].apply(lambda x: list(set(x)))
df_TA_view.head(2)
```

```python
df_TA_view["len_P"]=df_TA_view["trans_P_zips"].apply(len)
df_TA_view["len_S"]=df_TA_view["trans_S_zips"].apply(len)
df_TA_view["len_all"]=df_TA_view["TA_zip_list"].apply(len)
df_TA_view.head(3)
```

```python
df_TA_view[df_TA_view['len_P']+df_TA_view['len_S']!=df_TA_view['len_all']]
```

```python
df_TA_view['max_overlap_TA']=None
df_TA_view['max_overlap_pctg']=None

for i, row in df_TA_view.iterrows():
    zip_i=row['TA_zip_list']
    max_overlap=0
    max_TA=None
    max_pctg=0
    for j in range(i+1,len(df_TA_view)):
        zip_j=df_TA_view.loc[j,"TA_zip_list"]
        intersection=list(set(zip_i).intersection(zip_j))
        pctg=len(intersection)/len(set(zip_i))
        if pctg>max_overlap:
            max_pctg=pctg
            max_TA=df_TA_view.loc[j,'TA_num']
    df_TA_view.loc[i,'max_overlap_TA']=max_TA
    df_TA_view.loc[i,'max_overlap_pctg']=max_pctg

    
            
            
```
