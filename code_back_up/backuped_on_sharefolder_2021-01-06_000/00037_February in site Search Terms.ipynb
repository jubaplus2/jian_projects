```python
All_High=pd.read_csv('/home/jian/Projects/Big_Lots/Weather/Internal_Search_Phrase/data/All_High.csv',dtype=str,skiprows=26)
All_High=All_High.iloc[:,[1,2,4]]
All_High.columns=['Search_Phrase','All_Searches','All_High_Searches']

All_Low=pd.read_csv('/home/jian/Projects/Big_Lots/Weather/Internal_Search_Phrase/data/All_Low.csv',dtype=str,skiprows=26)
All_Low=All_Low.iloc[:,[1,2,4]]
All_Low.columns=['Search_Phrase','All_Searches','All_Low_Searches']
```

```python
Organic_High=pd.read_csv('/home/jian/Projects/Big_Lots/Weather/Internal_Search_Phrase/data/Organic_High.csv',dtype=str,skiprows=26)
Organic_High=Organic_High.iloc[:,[1,2,4]]
Organic_High.columns=['Search_Phrase','All_Searches','Organic_High_Searches']

Organic_Low=pd.read_csv('/home/jian/Projects/Big_Lots/Weather/Internal_Search_Phrase/data/Organic_Low.csv',dtype=str,skiprows=26)
Organic_Low=Organic_Low.iloc[:,[1,2,4]]
Organic_Low.columns=['Search_Phrase','All_Searches','Organic_Low_Searches']
```
