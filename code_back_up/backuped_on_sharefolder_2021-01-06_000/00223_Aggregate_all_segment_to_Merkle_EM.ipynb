```python
import pandas as pd
import datetime
import glob
import os

print(datetime.datetime.now())
print(os.getcwd())
```

```python
folder_aggregated_file="/mnt/clients/juba/hqjubaapp02/sharefolder/GoodYear/jian/Audience_selections/output_aggregated_file/"
try:
    os.stat(folder_aggregated_file)
except:
    os.mkdir(folder_aggregated_file)
    
list_all_segment_files=glob.glob("/mnt/clients/juba/hqjubaapp02/sharefolder/GoodYear/jian/Audience_selections/output_cleaned_audience_2020-11-06/*.csv")
print(len(list_all_segment_files))
```

```python
df_all_selected_ids=pd.DataFrame()
for file in list_all_segment_files:
    try:
        df=pd.read_csv(file,dtype=str,usecols=['indiv_key','best_email_txt','segment_name'])
    except:
        df=pd.read_csv(file,dtype=str,usecols=['indiv_key','best_email_txt'])
        df['segment_name']=os.path.basename(file).split(".")[0]
        
    df_all_selected_ids=df_all_selected_ids.append(df)
df_all_selected_ids['TestOrControl']=df_all_selected_ids['segment_name'].apply(lambda x: x.split("_")[1])
```

```python
df_all_selected_ids.to_csv(folder_aggregated_file+"GY_all_defined_CRM_segment_JL_%s.csv"%str(datetime.datetime.now().date()),index=False)
```
