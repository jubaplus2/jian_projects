```python
import pandas as pd
import os
import glob
import datetime
import random
import numpy as np
pd.options.display.max_columns=30
os.getcwd()
```

```python
df_in_18=pd.read_csv("./output_2019-10-20/BL_2019Q4_Final_Audience_details_18mo_shoppers_with_emails_seg_JL_2019-10-23.csv",
                     usecols=['customer_id_hashed',"email_address_hash","frmindex","segment","Recency_Status","HML_Group","Density_Tier","SOTF","ExperianCluster","zip_type"],dtype=str)
df_in_18.head(2)
```

```python
random.seed(1)

total_rows=len(df_in_18)

list_test_all_regular=[]
list_control_all_regular=[]

i_counter=0

for seg,group in df_in_18.groupby(['segment','frmindex']):
    random_list=random.sample(range(len(group)), int(np.round(len(group)/total_rows*500000)))

    group=group.reset_index()
    del group['index']
    group=group.reset_index()
    df_control=group[group['index'].isin(random_list)]
    df_test=group[~group['index'].isin(random_list)]
    
    df_control['TestGroup']="C"
    df_test['TestGroup']="T"
    
    df_control['segment_2019Q4']="C_"+df_control['segment']+"_RFM"
    df_test['segment_2019Q4']="T_"+df_test['segment']+"_RFM"
    
    list_control_all_regular.append(df_control)
    list_test_all_regular.append(df_test)
    
    i_counter+=1
    if i_counter%10==1:
        print(i_counter,datetime.datetime.now())
    
df_test_all_regular=pd.concat(list_test_all_regular)
df_control_all_regular=pd.concat(list_control_all_regular)
```

```python
df_19_to_48=pd.read_csv("./output_2019-10-20/BL_2019Q4_Final_Audience_details_19_to_48_Lapsed_shoppers_with_emails_seg_JL_2019-10-23.csv"
                        ,dtype=str)
df_19_to_48.head(2)
```

```python
random.seed(2)
total_rows=len(df_19_to_48)

list_test_all_19_to_48=[]
list_control_all_19_to_48=[]

i_counter=0

for seg,group in df_19_to_48.groupby(['segment']):
    random_list=random.sample(range(len(group)), int(np.round(len(group)/total_rows*100000)))

    group=group.reset_index()
    del group['index']
    group=group.reset_index()
    df_control=group[group['index'].isin(random_list)]
    df_test=group[~group['index'].isin(random_list)]
    
    df_control['TestGroup']="C"
    df_test['TestGroup']="T"
    
    df_control['segment_2019Q4']="C_"+df_control['segment']+"_Lapsed"
    df_test['segment_2019Q4']="T_"+df_test['segment']+"_Lapsed"
    
    list_control_all_19_to_48.append(df_control)
    list_test_all_19_to_48.append(df_test)
    
    i_counter+=1
    print(i_counter,datetime.datetime.now())
    
df_test_all_19_to_48=pd.concat(list_test_all_19_to_48)
df_control_all_19_to_48=pd.concat(list_control_all_19_to_48)
```

```python
write_folder_uploaded="/home/jian/Projects/Big_Lots/Live_Ramp/Quarterly_Update_2019Q4/uploaded_no_T_zip/"
write_folder_all_needed_audience="/home/jian/Projects/Big_Lots/Live_Ramp/Quarterly_Update_2019Q4/all_audiences/"

try:
    os.stat(write_folder_uploaded)
except:
    os.mkdir(write_folder_uploaded)
    
try:
    os.stat(write_folder_all_needed_audience)
except:
    os.mkdir(write_folder_all_needed_audience)
```

```python
df_test_all_regular.head(3)
```

```python
df_test_all_19_to_48['segment_2019Q4'].unique()
```

```python
df_summary=pd.DataFrame(columns=['segment','id_count'])

for seg,df_group in df_test_all_regular.groupby("segment_2019Q4"):    
    if "_T_2019Q4" not in seg:
        df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
        df.to_csv(write_folder_uploaded+seg.replace(" ","")+".csv",index=False)
    df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
    row_count=df.shape[0]
    df_summary=df_summary.append(pd.DataFrame({"segment":[seg],"id_count":[row_count]},index=["Test_in_18"]))
    df.to_csv(write_folder_all_needed_audience+seg.replace(" ","")+".csv",index=False)
        
        
for seg,df_group in df_control_all_regular.groupby("segment_2019Q4"):    
    if "_T_2019Q4" not in seg:
        df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
        df.to_csv(write_folder_uploaded+seg.replace(" ","")+".csv",index=False)
    df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
    row_count=df.shape[0]
    df_summary=df_summary.append(pd.DataFrame({"segment":[seg],"id_count":[row_count]},index=["Control_in_18"]))
    df.to_csv(write_folder_all_needed_audience+seg.replace(" ","")+".csv",index=False)   
    
for seg,df_group in df_test_all_19_to_48.groupby("segment_2019Q4"):    
    
    df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
    row_count=df.shape[0]
    df_summary=df_summary.append(pd.DataFrame({"segment":[seg],"id_count":[row_count]},index=["Test_19to48"]))
    df.to_csv(write_folder_uploaded+seg.replace(" ","")+".csv",index=False)
    df.to_csv(write_folder_all_needed_audience+seg.replace(" ","")+".csv",index=False)
        
        
for seg,df_group in df_control_all_19_to_48.groupby("segment_2019Q4"):    
    df=df_group[['customer_id_hashed','email_address_hash','segment_2019Q4']].rename(columns={"segment_2019Q4":"segment"})
    row_count=df.shape[0]
    df_summary=df_summary.append(pd.DataFrame({"segment":[seg],"id_count":[row_count]},index=["Control_19to48"]))
    df.to_csv(write_folder_uploaded+seg.replace(" ","")+".csv",index=False)
    df.to_csv(write_folder_all_needed_audience+seg.replace(" ","")+".csv",index=False)
```
