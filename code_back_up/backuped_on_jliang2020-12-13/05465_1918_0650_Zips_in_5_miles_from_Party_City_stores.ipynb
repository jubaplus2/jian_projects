```python
output_lat_long=Party_City_Address_Same.append(Party_City_Address_Diff_Same).append(Party_City_Address_Diff_Diff)
output_lat_long=output_lat_long.reset_index()
del output_lat_long['index']
```

```python
output_by_zip=pd.DataFrame()
output_lat_long['zip_list_5_miles']=np.nan
for i in range(len(output_lat_long)):
    store_center=(output_lat_long['google_lat'][i],output_lat_long['google_lng'][i])
    list_zip_5_miles=[output_lat_long['Ship Zip'][i],output_lat_long['google_zip'][i]]
    df=pd.DataFrame({"store":output_lat_long['Location'][i],"zip":list_zip_5_miles,"distance_mile":"zip_of_store"},index=[0,1]).drop_duplicates()
    for zip_code in zip_centers.keys():
        dist=haversine(store_center,zip_centers[zip_code],miles=True)
        if dist<=5:
            list_zip_5_miles=list_zip_5_miles+[zip_code]
            df=df.append(pd.DataFrame({"store":output_lat_long['Location'][i],"zip":zip_code,"distance_mile":dist},index=[0]))
    output_by_zip=output_by_zip.append(df)
    output_lat_long['zip_list_5_miles'][i]=[set(list_zip_5_miles)]
```

```python
df_unique_zip=output_by_zip[['zip']].drop_duplicates().rename(columns={"zip":"zips_in_5_miles"})
```
