# 2. create a dict, key: zip code, value: temp

```python
d = dict()
for k, v in response.iteritems():
    d[k] = v['main']['temp']
```

```python
d
```

# 3. filter zips  270 < temp < 280

```python
for k in d.keys():
    if d[k] <= 270 or d[k] >= 280:
        del d[k]
```

# 4. print date and sum of all zips' temp in all files under folder jsons
# eg. '2017-12-21 405008.848'

```python
import os
files = os.listdir('jsons/')
```

```python
result = dict()
for f in files:
    date = f.split('.')[0]
    with open('jsons/%s'%f, 'r') as foo:
        response = json.load(foo)
    print date, sum([x['main']['temp'] for x in response.values()])
```
