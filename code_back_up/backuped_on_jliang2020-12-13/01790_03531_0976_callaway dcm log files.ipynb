```python
def extract_event_time(df):
    event_time = df['Event Time'].apply(lambda x: datetime.datetime.fromtimestamp(x/1000000))

    df['event_date'] = event_time.apply(lambda x: x.strftime('%Y-%m-%d'))
    '''
    df = df[df['event_date'] >= '2017-04-02'].copy()

    df['event_time'] = event_time.apply(lambda x: x.strftime('%H:%M'))

    def daypart(s):
        if s < '06:00':
            return '00:00 to 06:00'
        elif s < '12:00':
            return '06:00 to 12:00'
        elif s < '18:00':
            return '12:00 to 18:00'
        else:
            return '18:00 to 00:00'
    df['event_daypart'] = df['event_time'].apply(daypart)
    '''
    weekday = '1_monday 2_tuesday 3_wednesday 4_thursday 5_friday 6_saturday 7_sunday'.split()

    df['event_weekday'] = event_time.apply(lambda x: weekday[x.weekday()])
    '''
    def get_week(dt):
        day1 = datetime.datetime(2017, 4, 2)
        n = (dt - day1).days // 7
        return 'week_of_'+(day1 + datetime.timedelta(7*n)).strftime('%Y-%m-%d')
    df['week_from_0402'] = event_time.apply(get_week)
    '''
    return df
```

## groups

```python
def group_sum(l):
    s = 0
    for x in l:
        s += x
    return s
activity_by_id = activity.groupby('User ID')['Total Conversions'].apply(group_sum)
activity_by_id.name = 'conversions'
```

```python
columns = ['OS', 'Site (DCM)', 'Creative', 'Creative Type', 'Creative Pixel Size', 
           'week_from_0402', 'event_weekday', 'event_daypart', 'Event Sub-Type']
conversion_by_id_by_c = dict()
for c in columns:
    df = activity.groupby(['User ID', c])['Total Conversions'].apply(group_sum)
    df = df.unstack().join(activity_by_id).sort_values('conversions', ascending=False)
    conversion_by_id_by_c[c] = df
    df.to_csv('/home/jubauser1/jzou/dcm_account7252/activity_groups/conversion_by_id_by_%s.csv'%c)
```

```python
columns = ['conversion_from_Callaway - Thank You Page Transactions', 
           'conversion_from_Callaway - Thank You Page Items Sold']
for c in columns:
    df = activity.groupby(['User ID', 'Site (DCM)'])[c].apply(group_sum)
    df = df.unstack().join(activity_by_id).sort_values('conversions', ascending=False)
    col_rename = {x:c[42:]+'_from_'+x for x in df.columns}
    df = df.rename(columns=col_rename)
    conversion_by_id_by_c[c] = df
    df.to_csv('/home/jubauser1/jzou/dcm_account7252/activity_groups/conversion_by_id_by_%s.csv'%c)
```

```python
activity_sorted = list(activity['Activity'].value_counts().sort_values(ascending=False).keys())
df = activity.groupby(['User ID', 'Activity'])['Total Conversions'].apply(group_sum)
df = df.unstack().join(activity_by_id).sort_values('conversions', ascending=False)[['conversions']+activity_sorted]
conversion_by_id_by_c['Activity'] = df
df.to_csv('/home/jubauser1/jzou/dcm_account7252/activity_groups/conversion_by_id_by_Activity.csv')
```

```python
keyword_sorted = list(activity['Paid Search Keyword'].value_counts().sort_values(ascending=False).keys())
df = activity.groupby(['User ID', 'Paid Search Keyword'])['Total Conversions'].apply(group_sum)
df = df.unstack().join(activity_by_id).sort_values('conversions', ascending=False)[['conversions']+keyword_sorted]
conversion_by_id_by_c['Paid Search Keyword'] = df
df.to_csv('/home/jubauser1/jzou/dcm_account7252/activity_groups/conversion_by_id_by_Paid Search Keyword.csv')
```
