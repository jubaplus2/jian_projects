```python
import glob
import pandas as pd
from sqlalchemy import create_engine  
```

```python
import logging
logging.basicConfig(filename='impr_to_sql.log', 
                    level=logging.INFO, 
                    format='%(asctime)s %(message)s')

import json
files_in_mysql = json.load(open('impr_files_in_mysql.json', 'r'))

files = glob.glob('/home/jubauser1/jzou/dcm_account7252/impression/*')
files = [x for x in files if x not in files_in_mysql]
files.sort()

engine = create_engine('mysql://root:JubaPlus-2017@localhost/dcm_account7252')

for f in files:
    try:
        df = pd.read_csv(f, index_col=['Campaign ID', 'Event Time'])
        df['file_name'] = f

        df.to_sql('impression', con=engine, if_exists='append')

        files_in_mysql.append(f)
        logging.info(f)
        del df
    except Exception as e:
        logging.warning('file: %s, warning:%s'%(f, e))

json.dump(files_in_mysql, open('impr_files_in_mysql.json', 'w'))
```
