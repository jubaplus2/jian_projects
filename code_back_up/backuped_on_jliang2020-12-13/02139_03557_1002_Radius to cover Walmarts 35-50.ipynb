```python
import pandas as pd
import numpy as np
from haversine import haversine
```

```python
markets_18_center=pd.read_excel("/home/jian/SunnyD/Walmart_Kroger_Publix/markets_center_18_23.xlsx")
```

```python
stores_3_brand=pd.read_csv("/home/jian/SunnyD/Walmart_Kroger_Publix/Jay_3_brands_zip_within_10mile.csv")
stores_lat_lng=stores_3_brand[['store_id','lat','lng']]
```

```python
output_all=pd.DataFrame()
output_qualified=pd.DataFrame()
for center in markets_18_center['Markets']:
    base_radius=60
    walmart_store_count=0
    
    df=markets_18_center[markets_18_center['Markets']==center]
    df.reset_index(inplace=True)
    del df['index']
        
    df_app=pd.DataFrame(columns=["Market","store","Store_Brand","Store_Lat","Store_Long","Dist","Cent_Lat","Cent_Long"])

    for store in stores_lat_lng['store_id']:
        df_store=stores_lat_lng[stores_lat_lng['store_id']==store]
        df_store.reset_index(inplace=True)
        del df_store['index']
        dist=haversine((df['Center_Lat'][0],df['Center_Long'][0]),(df_store['lat'][0],df_store['lng'][0]),miles=True)
        df_store=pd.DataFrame({"Market":center,"store":store,"Store_Brand":store.split("_")[0],
                                   "Store_Lat":df_store['lat'][0],"Store_Long":df_store['lng'][0],"Dist":dist,"Cent_Lat":df['Center_Lat'][0],"Cent_Long":df['Center_Long'][0]},index=[0])
        df_app=df_app.append(df_store)
    df_app.reset_index(inplace=True)
    del df_app['index']
    output_all=output_all.append(df_app)
    
    
    while ((walmart_store_count>50)|(walmart_store_count<35)):
        
        walmart_store_count=len(df_app[(df_app['Store_Brand']=="walmart")&(df_app['Dist']<=base_radius)])
        
        print("Center: "+center+" | Walmarts: "+str(walmart_store_count)+" | Radius: "+str(base_radius))
        if walmart_store_count<35:
            base_radius=base_radius+5
        elif walmart_store_count>50:
            base_radius=base_radius-5 
        else:
            qualified_df=df_app[df_app['Dist']<=base_radius]
            qualified_df['radius']=base_radius
            qualified_df['radius']['walmarkts_count']=walmart_store_count
    
    output_qualified=output_qualified.append(qualified_df)
    print (center)
    
    
        
        
```

```python
writer=pd.ExcelWriter("/home/jian/SunnyD/Walmart_Kroger_Publix/stores_in_radius.xlsx")
```

```python
output_qualified.to_excel(writer,"qualified in radius",index=False)
output_all.to_excel(writer,"data",index=False)
```

```python
writer.save()
```
