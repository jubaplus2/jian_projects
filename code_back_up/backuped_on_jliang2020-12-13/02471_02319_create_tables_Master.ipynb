```python
# DB name: BigLots
BL_SQL_CONNECTION= 'mysql+pymysql://jian:JubaPlus-2017@localhost/BigLots'  # BL_Rewards_Master
BL_engine = create_engine(
        BL_SQL_CONNECTION, 
        pool_recycle=1800
    )

existing_tables=pd.read_sql("show tables;",con=BL_engine)
existing_tables
```

```python
import pymysql.cursors
engine_pymysql_db_BL = pymysql.connect(host='localhost',user='jian',
                         password='JubaPlus-2017',db='BigLots',
                         charset='utf8mb4',cursorclass=pymysql.cursors.DictCursor)
```

```python
def create_BL_Master_table():
    with engine_pymysql_db_BL.cursor() as cur:
        cur.execute("DROP TABLE IF EXISTS BL_Rewards_Master")
        cur.execute("CREATE TABLE BL_Rewards_Master \
        (\
        customer_id_hashed char(64), \
        email_address_hash char(64), \
        sign_up_date Date, \
        sign_up_channel varchar(64), \
        sign_up_location int, \
        customer_zip_code varchar(16), \
        transaction_count int, \
        transaction_amount decimal(10,2), \
        experian_multi_cluster varchar(32), \
        experian_demo_cluster varchar(32), \
        purchase_channel varchar(32), \
        email_unsubscribe_indicator varchar(2), \
        email_undeliverable_indicator varchar(2), \
        file_path varchar(256) \
        );")
    print("1:An empty TABLE BL_Rewards_Master has been created.",datetime.datetime.now())
 
 
create_BL_Master_table()
```

```python
df_all_master.to_sql("BL_Rewards_Master",if_exists='replace', con=BL_engine, index=False,chunksize=300000,
                    dtype={
                        'customer_id_hashed':sqlalchemy.types.VARCHAR(length=64),
                        'email_address_hash':sqlalchemy.types.VARCHAR(length=64),
                        'sign_up_date':sqlalchemy.Date(), 
                        'sign_up_channel':sqlalchemy.types.VARCHAR(length=64),
                        'sign_up_location':sqlalchemy.types.INTEGER(),
                        'customer_zip_code':sqlalchemy.types.VARCHAR(length=16),
                        'transaction_count':sqlalchemy.types.INTEGER(),
                        'transaction_amount':sqlalchemy.types.DECIMAL(precision=10,scale=2,asdecimal=True),
                        'experian_multi_cluster':sqlalchemy.types.VARCHAR(length=32),
                        'experian_demo_cluster':sqlalchemy.types.VARCHAR(length=32),
                        'purchase_channel':sqlalchemy.types.VARCHAR(length=32),
                        'email_unsubscribe_indicator':sqlalchemy.types.VARCHAR(length=2),
                        'email_undeliverable_indicator':sqlalchemy.types.VARCHAR(length=2),
                        'file_path':sqlalchemy.types.VARCHAR(length=256)
                    })
                           
```
