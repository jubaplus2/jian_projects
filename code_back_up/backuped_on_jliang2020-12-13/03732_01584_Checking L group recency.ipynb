```python
df_0_18_detail=pd.read_csv("./BL_2020Q1_0_to_18_details_JL_2020-01-27.csv",dtype=str,usecols=['customer_id_hashed','email_address_hash','last_transaction_date','HML_Group',"zip_type"])
print(df_0_18_detail.shape)
df_0_18_detail.head(2)
```

```python
df_0_18_detail_without_T=df_0_18_detail[df_0_18_detail['zip_type']!="T"]
print(df_0_18_detail_without_T['zip_type'].unique())
```

```python
df_group_last_trans_without_T=df_0_18_detail_without_T.groupby(["HML_Group","last_transaction_date"])['customer_id_hashed'].count().to_frame().reset_index()

df_group_last_trans_wide_without_T=df_group_last_trans_without_T.pivot_table(index="HML_Group",columns="last_transaction_date",values="customer_id_hashed")
df_group_last_trans_wide_without_T=df_group_last_trans_wide_without_T.fillna(0)

writer=pd.ExcelWriter("./BL_QC_2020Q1_recency_by_RFM_group_without_Tzip_JL_"+str(datetime.datetime.now().date())+".xlsx",engine="xlsxwriter")
df_group_last_trans_wide_without_T.to_excel(writer,"wide_table_no_T")
df_group_last_trans_without_T.to_excel(writer,"long_table_no_T")
writer.save()
```

```python
df_group_last_trans=df_0_18_detail.groupby(["HML_Group","last_transaction_date"])['customer_id_hashed'].count().to_frame().reset_index()
```

```python
df_group_last_trans_wide=df_group_last_trans.pivot_table(index="HML_Group",columns="last_transaction_date",values="customer_id_hashed")
df_group_last_trans_wide=df_group_last_trans_wide.fillna(0)
```

```python
writer=pd.ExcelWriter("./BL_QC_2020Q1_recency_by_RFM_group_JL_"+str(datetime.datetime.now().date())+".xlsx",engine="xlsxwriter")
df_group_last_trans_wide.to_excel(writer,"wide_table")
df_group_last_trans.to_excel(writer,"long_table")
writer.save()
```
