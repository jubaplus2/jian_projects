```python
import pandas as pd
import os
import numpy as np
os.getcwd()
```

```python
def trim_tailing_space(x):
    if x is None:
        y=x
    else:
        y=x.rstrip()
    return y
```

```python
data=pd.read_table('/home/jian/Projects/C21/c21stores_Products_2019012813.txt.xz',
                 compression="xz",dtype=str,sep="|",lineterminator="\n",warn_bad_lines=True,error_bad_lines=False)
data.columns=[x.rstrip() for x in data.columns.tolist()]
data=data.fillna("Filled_None")
for col in data.columns.tolist():
    data[col]=data[col].apply(trim_tailing_space)
data=data.replace("Filled_None",np.nan)
print(data.shape)
data=data.drop_duplicates()
data.to_csv("/home/jian/Projects/C21/deduped_products_info.csv",index=False)
print(data.shape)
```
