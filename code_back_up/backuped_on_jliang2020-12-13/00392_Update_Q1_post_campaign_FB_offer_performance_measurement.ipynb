```python
df_prod_taxonomy=pd.read_table("/home/jian/BigLots/static_files/ProductTaxonomy/MediaStormProductTaxonomy20200501-152045-636.txt",
                               dtype=str,sep="|")
df_prod_taxonomy=df_prod_taxonomy[['division_id','class_code_id','subclass_id']].drop_duplicates()
```

```python
df_all_pos_5_weeks_rewards=pd.DataFrame()
# df_all_pos_5_weeks_nonrewards=pd.DataFrame()

i_counter=0
for file in list_pos_sales:
    df=pd.read_csv(file,dtype=str,sep="|")
    week_end_dt=df['transaction_dt'].max()
    df['week_end_dt']=week_end_dt
    
    df['item_transaction_amt']=df['item_transaction_amt'].astype(float)
    df['item_transaction_units']=df['item_transaction_units'].astype(int)

    df_rewards=df[pd.notnull(df['customer_id_hashed'])]
    # df_nonrewards=df[pd.isnull(df['customer_id_hashed'])]
        
    df_all_pos_5_weeks_rewards=df_all_pos_5_weeks_rewards.append(df_rewards)
    # df_all_pos_5_weeks_nonrewards=df_all_pos_5_weeks_nonrewards.append(df_nonrewards)
    
    i_counter+=1
    print(i_counter,datetime.datetime.now(),week_end_dt)
```

```python
print(df_all_pos_5_weeks_rewards.shape)
df_all_pos_5_weeks_rewards.head(2)
```

```python
df_all_pos_5_weeks_rewards=pd.merge(df_all_pos_5_weeks_rewards,df_prod_taxonomy,on=['class_code_id','subclass_id'],how="left")
df_all_pos_5_weeks_rewards['division_id']=df_all_pos_5_weeks_rewards['division_id'].fillna("-1")
df_all_pos_5_weeks_rewards.head(2)
```

```python
df_all_pos_5_weeks_rewards['store_type']=np.where(df_all_pos_5_weeks_rewards['location_id']=="6990","online","instore")
```

```python
df_all_pos_5_weeks_rewards_by_div=df_all_pos_5_weeks_rewards.groupby(["customer_id_hashed","store_type","division_id"])['item_transaction_amt'].sum().to_frame().reset_index()
df_all_pos_5_weeks_rewards_by_div_trans=df_all_pos_5_weeks_rewards[["customer_id_hashed","store_type","division_id","location_id","transaction_dt","transaction_id"]].drop_duplicates()
df_all_pos_5_weeks_rewards_by_div_trans=df_all_pos_5_weeks_rewards_by_div_trans.groupby(["customer_id_hashed","store_type","division_id"])['transaction_id'].count().to_frame().reset_index()

df_all_pos_5_weeks_rewards_by_div=pd.merge(df_all_pos_5_weeks_rewards_by_div,df_all_pos_5_weeks_rewards_by_div_trans,on=["customer_id_hashed","store_type","division_id"],how="outer")
df_all_pos_5_weeks_rewards_by_div=df_all_pos_5_weeks_rewards_by_div.fillna(0)

df_all_pos_5_weeks_rewards_by_div=df_all_pos_5_weeks_rewards_by_div.pivot_table(index=['customer_id_hashed'],columns=['division_id','store_type'],values=['item_transaction_amt','transaction_id'])
df_all_pos_5_weeks_rewards_by_div=df_all_pos_5_weeks_rewards_by_div.fillna(0)
df_all_pos_5_weeks_rewards_by_div.head(3)

```

```python
list_cols_new=[]
for col_tuple in df_all_pos_5_weeks_rewards_by_div.columns.tolist():
    col_new="_".join(col_tuple).replace("item_transaction_amt","sales").replace("transaction_id","trans")
    print(col_new,col_tuple)
    list_cols_new.append(col_new)
df_all_pos_5_weeks_rewards_by_div=df_all_pos_5_weeks_rewards_by_div.reset_index()
df_all_pos_5_weeks_rewards_by_div.columns=['customer_id_hashed']+list_cols_new

df_all_pos_5_weeks_rewards_by_div.head(2)
```

```python
df_seg_group=df_ID_matched_both[['customer_id_hashed','segment','group']].drop_duplicates()
print(df_seg_group[['customer_id_hashed','segment']].drop_duplicates().shape)
print(df_seg_group.shape,df_seg_group['customer_id_hashed'].nunique())

df_id_by_group=df_seg_group[['customer_id_hashed','segment']].drop_duplicates()

for group_name,df_group in df_seg_group.groupby("group"):
    df_group=df_group[['customer_id_hashed','group']].drop_duplicates()
    df_group=df_group.rename(columns={"group":group_name})
    df_group[group_name]=1
    df_id_by_group=pd.merge(df_id_by_group,df_group,on="customer_id_hashed",how="left")
df_id_by_group=df_id_by_group.fillna(0)
df_id_by_group.head(3)
```

```python
list_mobile_group=df_ID_matched_both['group'].unique().tolist()

cols_sum_agg=[x for x in df_id_by_group.columns.tolist() if "sales_" in x or "trans_" in x]
df_rewards_matched=pd.DataFrame()

for mobile_seg in list_mobile_group:
    df=df_id_by_group[df_id_by_group[mobile_seg]==1]
    df['sign_up_group']=np.where(pd.isnull(df['sign_up_date']),"old_rewards",
                                 np.where(df['sign_up_channel']=="ONLINE","new_online_rewards","new_instore_rewards")
                                )
    df_count_matched=df.groupby(['sign_up_group'])['customer_id_hashed'].count().to_frame().reset_index()
    df_sales_trans=df.groupby(['sign_up_group'])[cols_sum_agg].sum().reset_index()
    df=pd.merge(df_count_matched,df_sales_trans,on="sign_up_group",how="outer")
    df.insert(0,"group",mobile_seg)
    df_rewards_matched=df_rewards_matched.append(df)
df_rewards_matched=df_rewards_matched.rename(columns={"customer_id_hashed":"matched_id_count"})
```

```python
list_new_signed_id_online=df_id_by_group[df_id_by_group['sign_up_channel']=="ONLINE"]['customer_id_hashed'].unique().tolist()
list_new_signed_id_instore=df_id_by_group[df_id_by_group['sign_up_channel']=="STORE"]['customer_id_hashed'].unique().tolist()
print("len(list_new_signed_id_online)",len(list_new_signed_id_online))
print("len(list_new_signed_id_instore)",len(list_new_signed_id_instore))
```
