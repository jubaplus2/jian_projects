```python
# From Dom Store list
dict_SOTF={"5370":[27.3404981,-82.4994091],"5378":[41.6826177,-86.1878727]}
```

```python
output_df=pd.DataFrame()
for new_store in dict_SOTF.keys():
    i_counter=0
    store_center=dict_SOTF[new_store]
    for zip_cd in zip_centers.keys():
        dist=haversine(store_center,zip_centers[zip_cd],miles=True)
        if dist<=10:
            df=pd.DataFrame({"location_id":new_store,'store_lat':store_center[0],'store_long':store_center[1],"zip_cd":zip_cd,"revenue_flag":"new_in_10_miles","Distance (mile)":dist},index=[i_counter])
            i_counter+=1
            output_df=output_df.append(df)
output_df=output_df[['location_id','store_lat','store_long','zip_cd','revenue_flag','Distance (mile)']]
output_df.to_csv("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Additional_2_SOTF_zips_20190201/BL_zips_for_SOTF_5370_5378_JL_"+str(datetime.datetime.now().date())+".csv",index=False)

```
