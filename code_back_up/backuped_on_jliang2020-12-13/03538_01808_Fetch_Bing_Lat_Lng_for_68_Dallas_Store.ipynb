```python
SK_Dallas_Store=pd.read_excel('/home/jian/Projects/Smoothie_King/SK_BL_Store_Dallas_JL_2018-11-02.xlsx',sheetname="SK_Store_Dallas",dtype=str)
SK_Dallas_Store['zip']=SK_Dallas_Store['zip'].apply(lambda x: x.split("-")[0].zfill(5))
SK_Dallas_Store['search_key']=SK_Dallas_Store['Address']+", "+SK_Dallas_Store['city']+", "+SK_Dallas_Store['state']+", "+SK_Dallas_Store['zip']
SK_Dallas_Store['Bing_Lat']=np.nan
SK_Dallas_Store['Bing_Lng']=np.nan
SK_Dallas_Store['Bing_Address']=np.nan
for i in range(len(SK_Dallas_Store)):
    search_key=SK_Dallas_Store['search_key'][i]
    response=geocoder.bing(search_key,key="AuR1ORKdndGqHaD4jiql1C5Inwmc5BOqJrlo5joEAb2Yljp5aizGIN7IVoxJNcim")
    SK_Dallas_Store['Bing_Lat'][i]=response.latlng[0]
    SK_Dallas_Store['Bing_Lng'][i]=response.latlng[1]
    SK_Dallas_Store['Bing_Address'][i]=response.address
    
print(SK_Dallas_Store.shape)
SK_Dallas_Store.head(5)
```
