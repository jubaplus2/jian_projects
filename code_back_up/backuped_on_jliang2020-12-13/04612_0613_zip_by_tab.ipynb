```python
import pandas as pd
import os
import datetime
from haversine import haversine
import json
import numpy as np
import zipcodes

zip_centers=json.load(open("/home/jian/Docs/Geo_mapping/center_of_rentrak_zip.json","r"))

os.getcwd()
```

```python
tab_1=pd.read_excel("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Zips_by_Creative_Kayla_20190124/F and F contingency.xlsx",
                    sheetname="SATURDAY ONLY REG 2",dtype=str)

tab_2=pd.read_excel("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Zips_by_Creative_Kayla_20190124/F and F contingency.xlsx",
                    sheetname="Sat&Sun Reg 1 + Select Reg 5 +7",dtype=str)

tab_1=tab_1[['STORE','Address 1','City','ST','Zip']]
tab_1['Zip']=tab_1['Zip'].apply(lambda x: x.split("-")[0].zfill(5))
tab_2=tab_2[['STORE','Address 1','City','ST','Zip']]
tab_2['Zip']=tab_2['Zip'].apply(lambda x: x.split("-")[0].zfill(5))
```

```python
store_list=pd.read_table("/home/jian/BigLots/static_files/Store_list/MediaStormStores20190101-135843-638.txt",dtype=str,sep="|")
store_list=store_list[['location_id','address_line_1','city_nm','state_nm','zip_cd','latitude_meas','longitude_meas']]
store_list['zip_cd']=store_list['zip_cd'].apply(lambda x: x.split("-")[0].zfill(5))
store_list=store_list.rename(columns={"zip_cd":"store_zip5"})
```

```python
x=pd.merge(tab_1,store_list,left_on="STORE",right_on="location_id",how="left")

x[x['Zip']!=x['store_zip5']]
# The SFTP zips are more accurate based on Google map, and the store number matched
```

```python
x=pd.merge(tab_2,store_list,left_on="STORE",right_on="location_id",how="left")

x[x['Zip']!=x['store_zip5']]
# The SFTP zips are more accurate based on Google map, and the store number matched
```

```python
tab_1=store_list[store_list['location_id'].isin(tab_1['STORE'].unique())]
tab_2=store_list[store_list['location_id'].isin(tab_2['STORE'].unique())]

tab_1['latitude_meas']=tab_1['latitude_meas'].astype(float)
tab_1['longitude_meas']=tab_1['longitude_meas'].astype(float)

tab_2['latitude_meas']=tab_2['latitude_meas'].astype(float)
tab_2['longitude_meas']=tab_2['longitude_meas'].astype(float)
```

```python
store_level_zips=pd.read_csv("/home/jian/Projects/Big_Lots/New_TA/zips_in_new_ta/sales_by_zip (Store level).csv",dtype=str)
store_level_zips=store_level_zips[['location_id','zip','revenue_flag']].drop_duplicates()
store_level_zips=store_level_zips[store_level_zips['revenue_flag'].isin(['P','S'])]
```

```python
tab_1=pd.merge(tab_1,store_level_zips,on="location_id",how="left")
print(len(tab_1['location_id'].unique()))

tab_2=pd.merge(tab_2,store_level_zips,on="location_id",how="left")
print(len(tab_2['location_id'].unique()))

# All included in the old labeled stores, Easy
```

```python
tab_1['latitude_meas']=tab_1['latitude_meas'].astype(float)
tab_1['longitude_meas']=tab_1['longitude_meas'].astype(float)

tab_2['latitude_meas']=tab_2['latitude_meas'].astype(float)
tab_2['longitude_meas']=tab_2['longitude_meas'].astype(float)

tab_1=tab_1.reset_index()
del tab_1['index']

tab_2=tab_2.reset_index()
del tab_2['index']
```

```python
tab_1['dist_miles']=np.nan
for i in range(len(tab_1)):
    zipcode=tab_1['zip'][i]
    store_center_pair=(tab_1['latitude_meas'][i],tab_1['longitude_meas'][i])
    
    if zipcode in zip_centers.keys():
        zip_center_pair=zip_centers[zipcode]
        dist=haversine(zip_center_pair,store_center_pair,miles=True)
    else:
        try:
            zip_center_pair=zipcodes.matching(zipcode)
            dist=haversine(zip_center_pair,store_center_pair,miles=True)
        except:
            dist=np.nan
    tab_1['dist_miles'][i]=dist
            
```

```python
tab_2['dist_miles']=np.nan
for i in range(len(tab_2)):
    zipcode=tab_2['zip'][i]
    store_center_pair=(tab_2['latitude_meas'][i],tab_2['longitude_meas'][i])
    
    if zipcode in zip_centers.keys():
        zip_center_pair=zip_centers[zipcode]
        dist=haversine(zip_center_pair,store_center_pair,miles=True)
    else:
        try:
            zip_center_pair=zipcodes.matching(zipcode)
            dist=haversine(zip_center_pair,store_center_pair,miles=True)
        except:
            dist=np.nan
    tab_2['dist_miles'][i]=dist
            
```

```python
tab_1=tab_1[(tab_1['revenue_flag']=="P") | (tab_1['dist_miles']<=20)]
tab_2=tab_2[(tab_2['revenue_flag']=="P") | (tab_2['dist_miles']<=20)]

tab_1_unique_zips=tab_1[tab_1['revenue_flag']=="P"][['zip','revenue_flag']].append(tab_1[tab_1['revenue_flag']=="S"][['zip','revenue_flag']]).drop_duplicates('zip')
tab_2_unique_zips=tab_2[tab_2['revenue_flag']=="P"][['zip','revenue_flag']].append(tab_2[tab_2['revenue_flag']=="S"][['zip','revenue_flag']]).drop_duplicates('zip')
```

```python
writer=pd.ExcelWriter("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Zips_by_Creative_Kayla_20190124/BL_zips_by_creative_F_and_F_JL_"+str(datetime.datetime.now().date())+".xlsx",engine="xlsxwriter")
tab_1_unique_zips.to_excel(writer,"tab_1_unique_zips",index=False)
tab_2_unique_zips.to_excel(writer,"tab_2_unique_zips",index=False)
tab_1.to_excel(writer,"tab_1_zips_by_store",index=False)
tab_2.to_excel(writer,"tab_2_zips_by_store",index=False)
tab_1_input.to_excel(writer,"SATURDAY ONLY REG 2",index=False)
tab_2_input.to_excel(writer,"Sat&Sun Reg 1 + Select Reg 5 +7",index=False)
writer.save()
```
