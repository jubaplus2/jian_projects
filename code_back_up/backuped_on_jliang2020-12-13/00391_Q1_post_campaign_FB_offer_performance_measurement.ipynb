```python
def recursive_file_gen(root_path):
    for root, dirs, files in os.walk(root_path):
        for file in files:
            yield os.path.join(root,file)
```

```python
import sqlalchemy
BL_SQL_CONNECTION= 'mysql+pymysql://jian:JubaPlus-2017@localhost/BigLots' 
BL_engine = sqlalchemy.create_engine(
        BL_SQL_CONNECTION, 
        pool_recycle=1800
    )

df_all_rewards=pd.read_sql("select customer_id_hashed, sign_up_date from BL_Rewards_Master where sign_up_date>='2020-04-05';",con=BL_engine).drop_duplicates()
print(df_all_rewards.shape)
df_all_rewards.head(2)
```

```python
list_pos_sales=list(recursive_file_gen("/home/jian/BigLots/2020_by_weeks/"))
list_pos_sales=[x for x in list_pos_sales if "daily" in x.lower()]
list_pos_sales=[x for x in list_pos_sales if x[-4:]==".txt"]
# list_pos_sales=[x for x in list_pos_sales if x.split("MediaStorm_")[1][:10]<="2020-05-02"]
list_pos_sales=[x for x in list_pos_sales if x.split("MediaStorm_")[1][:10]>="2020-04-05"]
```

```python
list_pos_sales
```

```python
df_all_pos_5_weeks_rewards=pd.DataFrame()
df_all_pos_5_weeks_nonrewards=pd.DataFrame()

i_counter=0
for file in list_pos_sales:
    df=pd.read_csv(file,dtype=str,sep="|")
    week_end_dt=df['transaction_dt'].max()
    df['week_end_dt']=week_end_dt
    
    df['item_transaction_amt']=df['item_transaction_amt'].astype(float)
    df['item_transaction_units']=df['item_transaction_units'].astype(int)

    df_rewards=df[pd.notnull(df['customer_id_hashed'])]
    df_nonrewards=df[pd.isnull(df['customer_id_hashed'])]
        
    df_all_pos_5_weeks_rewards=df_all_pos_5_weeks_rewards.append(df_rewards)
    df_all_pos_5_weeks_nonrewards=df_all_pos_5_weeks_nonrewards.append(df_nonrewards)
    
    i_counter+=1
    print(i_counter,datetime.datetime.now(),week_end_dt)
```

```python
print(df_all_pos_5_weeks_rewards.shape)
df_all_pos_5_weeks_rewards.head(2)
```

```python
print(df_all_pos_5_weeks_nonrewards.shape)
df_all_pos_5_weeks_nonrewards.head(2)
```

```python
df_all_pos_5_weeks_rewards['store_type']=np.where(df_all_pos_5_weeks_rewards['location_id']=="6990","online","instore")
# 
df_all_pos_5_weeks_rewards_by_week=df_all_pos_5_weeks_rewards.groupby(["customer_id_hashed","store_type","week_end_dt"])['item_transaction_amt'].sum().to_frame().reset_index()
df_all_pos_5_weeks_rewards_by_week_trans=df_all_pos_5_weeks_rewards[["customer_id_hashed","store_type","week_end_dt","location_id","transaction_dt","transaction_id"]].drop_duplicates()
df_all_pos_5_weeks_rewards_by_week_trans=df_all_pos_5_weeks_rewards_by_week_trans.groupby(["customer_id_hashed","store_type","week_end_dt"])['transaction_id'].count().to_frame().reset_index()

df_all_pos_5_weeks_rewards_by_week=pd.merge(df_all_pos_5_weeks_rewards_by_week,df_all_pos_5_weeks_rewards_by_week_trans,on=["customer_id_hashed","store_type","week_end_dt"],how="outer")
df_all_pos_5_weeks_rewards_by_week=df_all_pos_5_weeks_rewards_by_week.fillna(0)

df_all_pos_5_weeks_rewards_by_week=df_all_pos_5_weeks_rewards_by_week.pivot_table(index=['customer_id_hashed'],columns=['week_end_dt','store_type'],values=['item_transaction_amt','transaction_id'])
df_all_pos_5_weeks_rewards_by_week=df_all_pos_5_weeks_rewards_by_week.fillna(0)


df_all_pos_5_weeks_rewards_by_week.head(2)
```

```python
list_cols_new=[]
for col_tuple in df_all_pos_5_weeks_rewards_by_week.columns.tolist():
    col_new="_".join(col_tuple).replace("item_transaction_amt","sales").replace("transaction_id","trans")
    print(col_new,col_tuple)
    list_cols_new.append(col_new)
df_all_pos_5_weeks_rewards_by_week=df_all_pos_5_weeks_rewards_by_week.reset_index()
df_all_pos_5_weeks_rewards_by_week.columns=['customer_id_hashed']+list_cols_new
```

```python
df_seg_group=df_ID_matched_both[['customer_id_hashed','segment','group']].drop_duplicates()
print(df_seg_group[['customer_id_hashed','segment']].drop_duplicates().shape)
print(df_seg_group.shape,df_seg_group['customer_id_hashed'].nunique())

df_id_by_group=df_seg_group[['customer_id_hashed','segment']].drop_duplicates()

for group_name,df_group in df_seg_group.groupby("group"):
    df_group=df_group[['customer_id_hashed','group']].drop_duplicates()
    df_group=df_group.rename(columns={"group":group_name})
    df_group[group_name]=1
    df_id_by_group=pd.merge(df_id_by_group,df_group,on="customer_id_hashed",how="left")
df_id_by_group=df_id_by_group.fillna(0)
```

```python
df_id_by_group=pd.merge(df_id_by_group,df_all_pos_5_weeks_rewards_by_week,on="customer_id_hashed",how="left")
df_id_by_group=df_id_by_group.fillna(0)
# 
df_id_by_group=pd.merge(df_all_rewards,df_id_by_group,on="customer_id_hashed",how="right")
```

```python
print(df_id_by_group[pd.notnull(df_id_by_group['sign_up_date'])]['sign_up_date'].min(),df_id_by_group[pd.notnull(df_id_by_group['sign_up_date'])]['sign_up_date'].max())
```

```python
list_mobile_group=df_ID_matched_both['group'].unique().tolist()

cols_sum_agg=[x for x in df_id_by_group.columns.tolist() if "-" in x]
df_rewards_matched=pd.DataFrame()

for mobile_seg in list_mobile_group:
    df=df_id_by_group[df_id_by_group[mobile_seg]==1]
    df['sign_up_group']=np.where(pd.isnull(df['sign_up_date']),"old_rewards","new_signed_rewards")
    df_count_matched=df.groupby(['sign_up_group'])['customer_id_hashed'].count().to_frame().reset_index()
    df_sales_trans=df.groupby(['sign_up_group'])[cols_sum_agg].sum().reset_index()
    df=pd.merge(df_count_matched,df_sales_trans,on="sign_up_group",how="outer")
    df.insert(0,"group",mobile_seg)
    df_rewards_matched=df_rewards_matched.append(df)
df_rewards_matched=df_rewards_matched.rename(columns={"customer_id_hashed":"matched_id_count"})
```

```python
list_new_signed_id=df_id_by_group[pd.notnull(df_id_by_group['sign_up_date'])]['customer_id_hashed'].unique().tolist()
df_all_pos_5_weeks_rewards['sign_up_group']=np.where(df_all_pos_5_weeks_rewards['customer_id_hashed'].isin(list_new_signed_id),"new_signed_rewards","old_rewards")

df_shopper_id_week_store_type_seg=df_all_pos_5_weeks_rewards[['customer_id_hashed','week_end_dt','store_type','sign_up_group']].drop_duplicates()
df_shopper_id_week_store_type_seg=pd.merge(df_shopper_id_week_store_type_seg,df_seg_group,on="customer_id_hashed",how="left")

df_shoppers_by_week_store_type=df_shopper_id_week_store_type_seg[['group','customer_id_hashed','week_end_dt','store_type','sign_up_group']]
df_shoppers_by_week_store_type=df_shoppers_by_week_store_type.groupby(['group','sign_up_group',"week_end_dt","store_type"])['customer_id_hashed'].nunique().to_frame().reset_index()

df_shoppers_by_week_both_storetype=df_shopper_id_week_store_type_seg[['group','customer_id_hashed','week_end_dt','sign_up_group']].drop_duplicates()
df_shoppers_by_week_both_storetype=df_shoppers_by_week_both_storetype.groupby(['group','sign_up_group',"week_end_dt"])['customer_id_hashed'].nunique().to_frame().reset_index()
```

```python
df_shoppers_by_week_store_type=df_shoppers_by_week_store_type.pivot_table(index=['group','sign_up_group'],columns=['week_end_dt','store_type'],values=['customer_id_hashed'])
df_shoppers_by_week_store_type=df_shoppers_by_week_store_type.fillna(0)
df_shoppers_by_week_store_type=df_shoppers_by_week_store_type.reset_index()
list_cols_new=[]
for col_tuple in df_shoppers_by_week_store_type.columns.tolist():
    col_new="_".join(col_tuple).replace("item_transaction_amt","sales").replace("transaction_id","trans").replace("__","")
    print(col_new,col_tuple)
    list_cols_new.append(col_new)
df_shoppers_by_week_store_type.columns=list_cols_new
```

```python
df_shoppers_by_week_both_storetype['store_type']="bothOnlineInstore"
df_shoppers_by_week_both_storetype=df_shoppers_by_week_both_storetype.pivot_table(index=['group','sign_up_group'],columns=['week_end_dt','store_type'],values=['customer_id_hashed'])
df_shoppers_by_week_both_storetype=df_shoppers_by_week_both_storetype.fillna(0)
df_shoppers_by_week_both_storetype=df_shoppers_by_week_both_storetype.reset_index()
list_cols_new=[]
for col_tuple in df_shoppers_by_week_both_storetype.columns.tolist():
    col_new="_".join(col_tuple).replace("item_transaction_amt","sales").replace("transaction_id","trans").replace("__","")
    print(col_new,col_tuple)
    list_cols_new.append(col_new)
df_shoppers_by_week_both_storetype.columns=list_cols_new
```

```python
df_shoppers_by_both_storetype_all_5_weeks=df_shopper_id_week_store_type_seg.groupby(["group",'sign_up_group'])['customer_id_hashed'].nunique().to_frame().reset_index()
df_shoppers_by_both_storetype_all_5_weeks=df_shoppers_by_both_storetype_all_5_weeks.rename(columns={"customer_id_hashed":"shopper_count_5_weeks"})

#
df_shoppers_by_both_storetype_first_4_weeks=df_shopper_id_week_store_type_seg[df_shopper_id_week_store_type_seg['week_end_dt']!="2020-05-09"]
df_shoppers_by_both_storetype_first_4_weeks=df_shoppers_by_both_storetype_first_4_weeks.groupby(["group",'sign_up_group'])['customer_id_hashed'].nunique().to_frame().reset_index()
df_shoppers_by_both_storetype_first_4_weeks=df_shoppers_by_both_storetype_first_4_weeks.rename(columns={"customer_id_hashed":"shopper_count_first_4_weeks"})

#
df_shoppers_by_both_storetype_mid_3_weeks=df_shopper_id_week_store_type_seg[df_shopper_id_week_store_type_seg['week_end_dt']!="2020-04-11"]
df_shoppers_by_both_storetype_mid_3_weeks=df_shoppers_by_both_storetype_mid_3_weeks[df_shoppers_by_both_storetype_mid_3_weeks['week_end_dt']!="2020-05-09"]
df_shoppers_by_both_storetype_mid_3_weeks=df_shoppers_by_both_storetype_mid_3_weeks.groupby(["group",'sign_up_group'])['customer_id_hashed'].nunique().to_frame().reset_index()
df_shoppers_by_both_storetype_mid_3_weeks=df_shoppers_by_both_storetype_mid_3_weeks.rename(columns={"customer_id_hashed":"shopper_count_mid_3_weeks"})

#
df_shoppers_by_both_storetype_last_4_weeks=df_shopper_id_week_store_type_seg[df_shopper_id_week_store_type_seg['week_end_dt']!="2020-04-11"]
df_shoppers_by_both_storetype_last_4_weeks=df_shoppers_by_both_storetype_last_4_weeks.groupby(["group",'sign_up_group'])['customer_id_hashed'].nunique().to_frame().reset_index()
df_shoppers_by_both_storetype_last_4_weeks=df_shoppers_by_both_storetype_last_4_weeks.rename(columns={"customer_id_hashed":"shopper_count_last_4_weeks"})


df_agg_shopper_count=pd.merge(df_shoppers_by_both_storetype_all_5_weeks,df_shoppers_by_both_storetype_first_4_weeks,on=["group",'sign_up_group'],how="outer")
df_agg_shopper_count=pd.merge(df_agg_shopper_count,df_shoppers_by_both_storetype_mid_3_weeks,on=["group",'sign_up_group'],how="outer")
df_agg_shopper_count=pd.merge(df_agg_shopper_count,df_shoppers_by_both_storetype_last_4_weeks,on=["group",'sign_up_group'],how="outer")
```

```python
df_output=pd.merge(df_rewards_matched,df_agg_shopper_count,on=["group",'sign_up_group'])
df_output
```
