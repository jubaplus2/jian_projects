```python
df_dma=df_dma.parse("2016_NSI_DMABYZIP (2)",dtype=str,skiprows=1)
df_dma=df_dma.iloc[:,[0,2,6]]
df_dma.columns=['zip_cd','dma','st']
```

```python
df_zip_all_info=pd.DataFrame()
for z in zip_centers.keys():
    df_from_nielsen=df_dma[df_dma['zip_cd']==z]
    dma=df_from_nielsen['dma'].unique().tolist()
    if len(dma)==0:
        response_list=zipcodes.matching(z)
        st_list=[]
        ct_list=[]
        
        for l in response_list:
            st_list.append(l['state'])
            ct_list.append(l['city'])
        st_list=list(set(st_list))
        ct_list=list(set(ct_list))
    else:
        st_list=df_from_nielsen['st'].unique().tolist()
        response_list=zipcodes.matching(z)
        ct_list=[]
        for l in response_list:
            ct_list.append(l['city'])
        ct_list=list(set(ct_list))
    df=pd.DataFrame({"zip_cd":z,
                     "city_list":str(ct_list),
                     "st_list":str(st_list),
                     "dma_list":str(dma)},index=[0])
    df_zip_all_info=df_zip_all_info.append(df)
df_zip_all_info.head(3)
```
