```python
import pandas as pd
import os
import numpy as np
import datetime
```

```python
def recrusive_file_gen(root_path):
    for root, dirs, files in os.walk(root_path):
        for file in files:
            yield os.path.join(root, file)
```

```python
all_2019_weekly_files=list(recrusive_file_gen("/home/jian/BigLots/2019_by_weeks/"))
all_2019_weekly_files=[x for x in all_2019_weekly_files if "Daily" in x]

all_2019_weekly_files_item_level=[]
for file in all_2019_weekly_files:
    df=pd.read_csv(file,nrows=10,dtype=str,sep="|")
    if "item_id" in df.columns.tolist():
        all_2019_weekly_files_item_level=all_2019_weekly_files_item_level+[file]
all_2019_weekly_files_item_level        
```

```python
output_price_0216_to_0406=pd.DataFrame()
i_counter=0
for file in all_2019_weekly_files_item_level:
    date=file.split("BigLots/2019_by_weeks/MediaStorm_")[1][:10]
    df=pd.read_csv(file,dtype=str,sep="|",usecols=['class_code_id','subclass_id','item_id','item_transaction_units','item_transaction_amt'])
    df['item_transaction_units']=df['item_transaction_units'].astype(int)
    df['item_transaction_amt']=df['item_transaction_amt'].astype(float)
    df['item_price']=df['item_transaction_amt']/df['item_transaction_units']
    df=df.groupby(['class_code_id','subclass_id','item_id','item_price'])['item_transaction_units'].count().to_frame().reset_index()
    df=df.rename(columns={"item_transaction_units":"transaction_record_counts"})
    # df=df[['class_code_id','subclass_id','item_id','item_price']].drop_duplicates()
    df['week_end_dt']=date
    output_price_0216_to_0406=output_price_0216_to_0406.append(df)
    i_counter+=1
    print(i_counter,datetime.datetime.now())
```

```python
output_price_0216_to_0406=output_price_0216_to_0406[['week_end_dt','item_id','class_code_id','subclass_id','item_price','transaction_record_counts']]
```
