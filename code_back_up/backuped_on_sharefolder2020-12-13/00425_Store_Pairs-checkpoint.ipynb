```python
query="SELECT * from mdb.vw__installer_master as t1 \
left join mdb_cdl.store t2 on t1.store_num=t2.store_num;"
print("query: \n%s"%query)
df_store_list_BQ_org = client.query(query).result().to_dataframe()

list_all_cols=df_store_list_BQ_org.columns.tolist()
list_cols_kept=['installer_store_num','installer_name','store_num','company_owned_store_ind','open_date','close_date',
           'address_1','city','state','zip_cd','country','latitude','longitude','timezone']
df_store_list_BQ=df_store_list_BQ_org[list_cols_kept]
df_store_list_BQ.head(5)
```

```python
# filter stores currently open only
print(df_store_list_BQ.shape)
df_store_list_BQ=df_store_list_BQ[pd.isnull(df_store_list_BQ['close_date'])]
print(df_store_list_BQ.shape)
```

```python
df_POI_brand=pd.read_table("/mnt/clients/juba/hqjubaapp02/sharefolder/Docs/POI/places_20200406/part-00000-tid-3169497161568699755-0a71d798-e636-4e2f-949e-02e4669f577c-43-c000.csv.gz",
                          compression="gzip",sep="|")
list_all_brands=df_POI_brand['brand_name'].unique().tolist()
list_GY_brands=[x for x in list_all_brands if "goodyear" in x.lower()]
list_GY_brands

list_GY_brands=['Goodyear Auto Service Centers', 'Goodyear Commercial Tire & Service Center']
```

```python
df_GY_locations=pd.DataFrame()
for file in list_poi_files:
    df=pd.read_csv(file,dtype=str)
    df_1=df[df['brands'].isin(list_GY_brands)]
    # df_2=df[df['location_name'].apply(lambda x: "good" in x.lower() and "year" in x.lower())]
    df_2=pd.DataFrame()
    df=df_1.append(df_2).drop_duplicates()
    df_GY_locations=df_GY_locations.append(df)
print(df_GY_locations.shape)
```
