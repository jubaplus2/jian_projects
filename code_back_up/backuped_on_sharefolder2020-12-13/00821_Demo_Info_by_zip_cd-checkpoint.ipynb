```python
import pandas as pd
import numpy as np
import datetime
import os
pd.options.display.max_columns=999
os.getcwd()
import gc
# Get teenagers and mothers only
```

```python
census_data_B01001=pd.read_csv("/home/jian/Docs/Household_and_Population/Cencus_2019/ACS_17_5YR_B01001_with_ann.csv",dtype=str,skiprows=1)
print(census_data_B01001.shape)
census_data_B01001.head(2)
```

```python
census_data_S0101=pd.read_csv("/home/jian/Docs/Household_and_Population/Cencus_2019/ACS_17_5YR_S0101_with_ann.csv",dtype=str,skiprows=1)
print(census_data_S0101.shape)
census_data_S0101.head(2)
```

```python
census_teenager=census_data_S0101[['Id','Id2','Geography','Total; Estimate; AGE - Under 5 years','Total; Estimate; AGE - 5 to 9 years',
                                   'Total; Estimate; AGE - 10 to 14 years','Total; Estimate; AGE - 15 to 19 years']]
census_teenager['Total; Estimate; AGE - Under 5 years']=census_teenager['Total; Estimate; AGE - Under 5 years'].astype(float)
census_teenager['Total; Estimate; AGE - 5 to 9 years']=census_teenager['Total; Estimate; AGE - 5 to 9 years'].astype(float)
census_teenager['Total; Estimate; AGE - 10 to 14 years']=census_teenager['Total; Estimate; AGE - 10 to 14 years'].astype(float)
census_teenager['Total; Estimate; AGE - 15 to 19 years']=census_teenager['Total; Estimate; AGE - 15 to 19 years'].astype(float)

census_teenager['teenagers_10_19']=census_teenager[['Total; Estimate; AGE - 10 to 14 years','Total; Estimate; AGE - 15 to 19 years']].sum(axis=1)
census_teenager=census_teenager.rename(columns={"Total; Estimate; AGE - Under 5 years":"kids_under_5"})
census_teenager=census_teenager.rename(columns={"Total; Estimate; AGE - 5 to 9 years":"kids_5_9"})
census_teenager=census_teenager[['Id2','kids_under_5','kids_5_9','teenagers_10_19']]
```

```python
census_female_25_54=census_data_B01001[['Id','Id2','Geography','Estimate; Female: - 25 to 29 years','Estimate; Female: - 30 to 34 years',
                                      'Estimate; Female: - 35 to 39 years','Estimate; Female: - 40 to 44 years',
                                      'Estimate; Female: - 45 to 49 years','Estimate; Female: - 50 to 54 years']]
for col in ['Estimate; Female: - 25 to 29 years','Estimate; Female: - 30 to 34 years',
                                      'Estimate; Female: - 35 to 39 years','Estimate; Female: - 40 to 44 years',
                                      'Estimate; Female: - 45 to 49 years','Estimate; Female: - 50 to 54 years']:
    census_female_25_54[col]=census_female_25_54[col].astype(float)

census_female_25_54['Female_25_54_pop']=census_female_25_54[['Estimate; Female: - 25 to 29 years','Estimate; Female: - 30 to 34 years',
                                                         'Estimate; Female: - 35 to 39 years','Estimate; Female: - 40 to 44 years',
                                                         'Estimate; Female: - 45 to 49 years','Estimate; Female: - 50 to 54 years']].sum(axis=1)

census_female_25_54=census_female_25_54[['Id2','Female_25_54_pop']]
```

```python
census_ZCTA_teen_mother=pd.merge(census_female_25_54,census_teenager,on="Id2",how="outer")
census_ZCTA_teen_mother=census_ZCTA_teen_mother.rename(columns={"Id2":"ZCTA"})
census_ZCTA_teen_mother.shape
```

```python
df_zip_ZCTA_mapping=pd.ExcelFile("/home/jian/Docs/Household_and_Population/Cencus_2019/zcta_zip_mapping/zip_to_zcta_2018.xlsx",dtype=str)
print(df_zip_ZCTA_mapping.sheet_names)
df_zip_ZCTA_mapping=df_zip_ZCTA_mapping.parse("ziptozcta2017",dtype=str)
print(df_zip_ZCTA_mapping.shape)
df_zip_ZCTA_mapping.head(2)

df_zip_ZCTA_mapping=df_zip_ZCTA_mapping[['ZIP_CODE',"ZCTA"]].drop_duplicates()
print(df_zip_ZCTA_mapping.shape)
print(df_zip_ZCTA_mapping['ZIP_CODE'].nunique())
```

```python
df_EASI_A=pd.read_csv("/home/jian/Docs/EASI/2018-07-19/ZIP4_18_DATA_A2_CSV/ZIP4_18_DATA_A2.CSV",
                      skiprows=1,usecols=['ZIP_CODE','ZIP4','POP18','HH18','AVGHHINC'])
df_EASI_A['ZIP_CODE']=df_EASI_A['ZIP_CODE'].astype(str).apply(lambda x: x.zfill(5))
df_EASI_A['ZIP4']=df_EASI_A['ZIP4'].astype(str).apply(lambda x: x.zfill(4))
df_EASI_A['total_income_HH_avg']=df_EASI_A['AVGHHINC']*df_EASI_A['HH18']

df_zip_income=df_EASI_A.groupby(["ZIP_CODE"])['total_income_HH_avg','HH18','POP18'].sum().reset_index()
df_zip_income['HH_inc_avg']=df_zip_income['total_income_HH_avg']/df_zip_income['HH18']
print(df_zip_income.shape)
df_zip_income.head(3)
```

```python
del df_EASI_A
gc.collect()
```

```python
df_EASI_C=pd.read_csv("/home/jian/Docs/EASI/2018-07-19/ZIP4_18_DATA_C2_CSV/ZIP4_18_DATA_C2.CSV",
                      skiprows=1,usecols=['ZIP_CODE','ZIP4','HH18','PCTHHL25','PCTHH25_34','PCTHH35_44','PCTHH45_54','PCTMARC','PCTSPARC'])
df_EASI_C['ZIP_CODE']=df_EASI_C['ZIP_CODE'].astype(str).apply(lambda x: x.zfill(5))
df_EASI_C['ZIP4']=df_EASI_C['ZIP4'].astype(str).apply(lambda x: x.zfill(4))

df_EASI_C['HH_0_25']=df_EASI_C['HH18']*df_EASI_C['PCTHHL25']/100
df_EASI_C['HH_25_34']=df_EASI_C['HH18']*df_EASI_C['PCTHH25_34']/100
df_EASI_C['HH_35_44']=df_EASI_C['HH18']*df_EASI_C['PCTHH35_44']/100
df_EASI_C['HH_45_54']=df_EASI_C['HH18']*df_EASI_C['PCTHH45_54']/100
df_EASI_C['HH_25_54']=df_EASI_C['HH_25_34']+df_EASI_C['HH_35_44']+df_EASI_C['HH_45_54']

df_EASI_C['HH_MARC']=df_EASI_C['HH18']*df_EASI_C['PCTMARC']/100
df_EASI_C['HH_SPARC']=df_EASI_C['HH18']*df_EASI_C['PCTSPARC']/100
df_EASI_C['HH_Children']=df_EASI_C['HH_MARC']+df_EASI_C['HH_SPARC']

df_zip_HH=df_EASI_C.groupby(["ZIP_CODE"])['HH18','HH_MARC','HH_SPARC','HH_Children','HH_25_54','HH_0_25'].sum().reset_index()
```

```python
del df_zip_income['HH18']
del df_zip_income['total_income_HH_avg']

df_zip_demo_EASI=pd.merge(df_zip_income,df_zip_HH,on="ZIP_CODE",how="outer")
pop_zip_18=df_zip_demo_EASI[['ZIP_CODE','POP18']]

mapping_HH_income=df_zip_demo_EASI[['ZIP_CODE','HH_inc_avg']]

df_zip_demo_EASI.head(2)
```

```python
df_zip_demo_census_output=pd.merge(census_ZCTA_teen_mother,df_zip_ZCTA_mapping,on="ZCTA",how="left")
df_zip_demo_census_output=pd.merge(df_zip_demo_census_output,pop_zip_18,on="ZIP_CODE",how="left")

zcta_pop=df_zip_demo_census_output.groupby(["ZCTA"])['POP18'].sum().to_frame().reset_index()
zcta_pop=zcta_pop.rename(columns={"POP18":"POP18_ZCTA"})

df_zip_demo_census_output=pd.merge(df_zip_demo_census_output,zcta_pop,on="ZCTA",how="left")
df_zip_demo_census_output['pop_pctg']=df_zip_demo_census_output['POP18']/df_zip_demo_census_output['POP18_ZCTA']

df_zip_demo_census_output['kids_under_5_census']=df_zip_demo_census_output['kids_under_5']*df_zip_demo_census_output['pop_pctg']
df_zip_demo_census_output['kids_5_9_census']=df_zip_demo_census_output['kids_5_9']*df_zip_demo_census_output['pop_pctg']

df_zip_demo_census_output['teenagers_10_19_census']=df_zip_demo_census_output['teenagers_10_19']*df_zip_demo_census_output['pop_pctg']
df_zip_demo_census_output['Female_25_54_pop_census']=df_zip_demo_census_output['Female_25_54_pop']*df_zip_demo_census_output['pop_pctg']

# df_zip_demo_census_output
```

```python
df_zip_demo_census_output=df_zip_demo_census_output[['ZIP_CODE','Female_25_54_pop_census','teenagers_10_19_census','kids_under_5_census','kids_5_9_census']]
df_zip_demo_census_output=df_zip_demo_census_output[pd.notnull(df_zip_demo_census_output['Female_25_54_pop_census'])]
```

```python
mapping_zipcd_child_pctg=df_zip_HH[['ZIP_CODE','HH18','HH_Children']]
mapping_zipcd_child_pctg['WithChild_Pctg']=df_zip_HH['HH_Children']/df_zip_HH['HH18']
mapping_zipcd_child_pctg=mapping_zipcd_child_pctg[['ZIP_CODE','WithChild_Pctg']]

##
df_zip_demo_census_output=pd.merge(df_zip_demo_census_output,mapping_zipcd_child_pctg,on="ZIP_CODE",how="left")
df_zip_demo_census_output['calculated_Monther_census']=df_zip_demo_census_output['Female_25_54_pop_census']*df_zip_demo_census_output['WithChild_Pctg']
```

```python
df_zip_demo_census_output=pd.merge(df_zip_demo_census_output,mapping_HH_income,on="ZIP_CODE",how="left")
```

```python
writer=pd.ExcelWriter("./SunnyD_US_zips_selected_demo_by_zip_JL_"+str(datetime.datetime.now().date())+".xlsx",engine="xlsxwriter")
df_zip_demo_census_output.to_excel(writer,"based_on_census",index=False)
writer.save()
```
