```python
def recursive_file_gen(root_path):
    for root, dirs, files in os.walk(root_path):
        for file in files:
            yield os.path.join(root,file)
list_all_files=list(recursive_file_gen("/home/jian/BigLots/"))
```

```python
list_all_POS=[x for x in list_all_files if "dailysales" in x.lower()]
list_all_POS=[x for x in list_all_POS if "by_weeks/MediaStorm_" in x]
list_2020_POS=[x for x in list_all_POS if x.split("by_weeks/MediaStorm_")[1][:10]>"2020-01-01"]


df_all_first_trans_by_id=pd.DataFrame()
for file in list_2020_POS:
    df=pd.read_tabel(file,usecols=['location_id','transaction_dt','transaction_id','customer_id_hashed'],dtype=str,sep="|")
    df=df.sort_values(['customer_id_hashed','transaction_dt','transaction_id','location_id']).drop_duplicates("customer_id_hashed")
    df_all_first_trans_by_id=df_all_first_trans_by_id.append(df)
    
df_all_first_trans_by_id=df_all_first_trans_by_id.sort_values(['customer_id_hashed','transaction_dt','transaction_id','location_id']).drop_duplicates("customer_id_hashed")
```
