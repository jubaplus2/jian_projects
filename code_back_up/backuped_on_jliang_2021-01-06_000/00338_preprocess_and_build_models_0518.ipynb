```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression 
```

```python
df_X=df_input_all_2[col_needed_X]
df_Y=df_input_all_2[col_needed_Y]
X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, test_size=0.25, random_state=0)
```

```python
# 1

clf_DV_1_week=LogisticRegression(random_state=0,solver='lbfgs', C=2,penalty='l2').fit(X_train.values.tolist(), y_train.iloc[:,0].values)
model = SelectFromModel(clf_DV_1_week, prefit=True)
X_train_new=model.transform(X_train)
X_test_new=X_test.loc[:,list(model.get_support())]

model_new_1=LogisticRegression(random_state=0,solver='lbfgs', C=2,penalty='l2').fit(X_train_new, y_train.iloc[:,0].values)
score = model_new_1.score(X_test_new.values.tolist(), y_test.iloc[:,0].values)
print(score)
```

```python
from sklearn import metrics
list_prob=[x[0] for x in model_new_1.predict_proba(X_test_new)]
threshold_list = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,.7,.75,.8,.85,.9,.95,.99]
for i in threshold_list:
    print ('\n******** For i = {} ******'.format(i))

    Y_test_pred=[1 if x<=i else 0 for x in list_prob]
    test_accuracy = metrics.accuracy_score(y_test.iloc[:,0],Y_test_pred)
    print(i,test_accuracy)
    
```

```python
metrics.accuracy_score(y_test.iloc[:,0],Y_test_pred)
```
