```python
last_week_start="'2019-12-22'"
last_week_end="'2019-12-28'"

df_lastweek_activities=pd.read_sql("select customer_id_hashed, count(distinct session_sequence) as lastweek_session from Pred_Activity_BL_id \
where date_utc between %s and %s group by customer_id_hashed;"%(last_week_start,last_week_end),con=BL_engine)
df_lastweek_activities.shape
```

```python
df_act_to_use=pd.merge(df_total_activities,df_lastweek_activities,on="customer_id_hashed",how="left")
df_act_to_use=df_act_to_use.fillna(0)
```
