```python
import pandas as pd
import numpy as np
import os
import datetime
samplerows=None
```

```python
chunksize_num = 10**7
filename='/home/jian/Projects/Big_Lots/Live_Ramp/Quarterly_Update_2019Q1/crm_newscore_0922/combinedtransactions_0922.csv'
dftrans_before_20180922=pd.DataFrame()
count_i=0

for chunk in pd.read_csv(filename, chunksize=chunksize_num,dtype=str,usecols=['transaction_date'],nrows=samplerows): #Add back the transaction info,,

    chunk = chunk.drop_duplicates()
    
    dftrans_before_20180922=dftrans_before_20180922.append(chunk)
    count_i+=1
    print(count_i,datetime.datetime.now())


del chunk
print("Earliest Date:" + str(dftrans_before_20180922['transaction_date'].min()))
print("Latest Date:" + str(dftrans_before_20180922['transaction_date'].max()))
```

```python
def recursive_file_gen(root_path):
    for root, dirs, files in os.walk(root_path):
        for file in files:
            yield os.path.join(root, file)

weekly_master_files_2019Q2_plus=list(recursive_file_gen("/home/jian/BigLots/2019_by_weeks/"))
weekly_master_files_2019Q2_plus=[x for x in weekly_master_files_2019Q2_plus if "aster" in x]
pd.read_csv(weekly_master_files_2019Q2_plus[2],dtype=str,nrows=10,sep="|")
```
