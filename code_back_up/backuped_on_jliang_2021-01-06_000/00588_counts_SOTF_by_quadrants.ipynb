```python
import pandas as pd
import os
import datetime
import numpy as np
```

```python
os.getcwd()
```

```python
SOTF_list=pd.ExcelFile("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Store_counts_SOTF_ocunts_by_Q4_Quadrants_Bruce_20190418/Store List Report 04.05.19 330PM.xlsx")
print(SOTF_list.sheet_names)
SOTF_list=SOTF_list.parse("Store List",dtype=str)
SOTF_list=SOTF_list.iloc[:1408,:]
```

```python
print(SOTF_list['SOTF'].unique().tolist())
SOTF_list=SOTF_list[SOTF_list['SOTF']=="Yes"]
SOTF_list['Open Date']=SOTF_list['Open Date'].apply(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d %H:%M:%S").date())
SOTF_list_before_Nov_1_2018=SOTF_list[SOTF_list['Open Date']<datetime.date(2018,11,1)]
SOTF_list_before_Nov_1_2018=SOTF_list_before_Nov_1_2018[['Store','SOTF']].rename(columns={"Store":"location_id"})
```

```python
Q4_quadrant_definations=pd.read_excel("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Store_counts_SOTF_ocunts_by_Q4_Quadrants_Bruce_20190418/Excel_BL_2018_Q4_post_YoY_small_JL_2019-03-04.xlsx",
                                      sheetname="Q4_Store_Quadrant_Defination",dtype=str,usecols=["location_id",'Quadrant'])
Q4_quadrant_definations.shape
```

```python
Q4_quadrant_definations=pd.merge(Q4_quadrant_definations,SOTF_list_before_Nov_1_2018,on="location_id",how="outer")
Q4_quadrant_definations['Quadrant']=Q4_quadrant_definations['Quadrant'].fillna("NotDefined")
Q4_quadrant_definations['SOTF']=Q4_quadrant_definations['SOTF'].fillna("No")
summary=Q4_quadrant_definations.groupby(['Quadrant','SOTF'])['location_id'].count().to_frame().reset_index()
```

```python
writer=pd.ExcelWriter("/home/jian/Projects/Big_Lots/Analysis/2019_Q1/Planner_Request/Store_counts_SOTF_ocunts_by_Q4_Quadrants_Bruce_20190418/BL_summary_SOTF_count_before_2018Nov_JL_20180418.xlsx",engine="xlsxwriter")
Q4_quadrant_definations.to_excel(writer,"Store_Quadrant_with_SOTF",index=False)
summary.to_excel(writer,"summary_1",index=False)
writer.save()
```
